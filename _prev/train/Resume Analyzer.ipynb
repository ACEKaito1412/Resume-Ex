```python
def get_spacy_doc(file, data):
    nlp = spacy.load('en_core_web_sm')
    db = DocBin()  # Create an instance of the DocBin class

    for text, annot in tqdm(data):
        doc = nlp.make_doc(text)
        annot = annot['entities']

        ents = []
        entity_indices = []

        for start, end, label in annot:
            # start = int(item['start'])
            # end = int(item['end'])
            # label = item['label']

            skip_entity = False

            for idx in range(start, end):
                if idx in entity_indices:
                    skip_entity = True
                    break

            if skip_entity:
                continue

            entity_indices = entity_indices + list(range(start, end))

            try:
                span = doc.char_span(start, end, label=label, alignment_mode='strict')

                if span is None:
                    err_data = str([start, end]) + "  " + str(text) + "\n"
                    file.write(err_data)
                else:
                    ents.append(span)

            except:
                continue

        try:
            doc.ents = ents
            db.add(doc)
        except:
            pass

    return db  # Return the completed DocBin object outside the loop
```

```python
import re

def remove_unnecessary_chars(text):
    # Remove special characters and symbols except ',', '.'
    resume_text = re.sub(r"[^a-zA-Z0-9\s,'.-@]", " ", text)

    # Remove HTML tags
    resume_text = re.sub(r"<.*?>", "", resume_text)

    # Remove non-printable characters
    resume_text = ''.join(filter(lambda x: x.isprintable(), resume_text))

    # Remove extra whitespaces and newlines
    resume_text = re.sub(r'\s+', ' ', resume_text)

    # Return the cleaned resume data
    return resume_text
```

```python
import PyPDF2

page_text = ""
# Open the PDF file in read-binary mode
with open('resume.pdf', 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)

    # Loop through each page and extract text
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        page_text += page.extract_text()

        print(f"Page {page_num + 1}:\n{page_text}\n")
```
